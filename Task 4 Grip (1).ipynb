```python
#loading necessary libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
#loading Dataset 
file="globalterrorism.csv"
raw_data=pd.read_csv(file)
raw_data.head()
```

```python
import chardet 
with open(file,'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result
```

# lets check dataset in more deep:


```python
data = raw_data.copy()
data.info()
```

```python
data.describe()
```

We can clearly see that mean of the countries with crime rate is approximately 132 and total is 181691


# Now check for any null values or insufficient data

```python
data.isna().sum()
```

# there so many null values but some columns are hidden so lets unhide them 

```python
nan_values = data.isna()
nan_columns = nan_values.any()

columns_with_nan = data.columns[nan_columns].tolist()
columns_with_nan
```

# lets explore and process all these columns one by one

```python
#renaming numerous column  for easy understandings 
data.rename(columns = {'iyear':'Year','imonth':'Month', 'iday':'Day', 'gname':'Group name', 'country_txt': 'Country','region_txt':'Region','attacktype1_txt':'AttackType',
                      'targtype1_txt':'TargetType','weaptype1_txt':'WeaponType','nwound':'Wound','nkill':'kills','provstate':'State'}, inplace=True)
```

```python
pd.set_option("display.max_columns",None)
print(data.dtypes)
```

# Note that all these columns are not necessary for data analysis 

So we take only those columns which are necessary for our analysis.


```python
data = data[['Year','Month','Day','Country','State','Region','city','latitude','longitude',"AttackType",'kills',
                'Wound','target1','summary','Group name','TargetType','WeaponType','motive']]
data.head()
             
```

```python
data.count()
```

```python
data.isnull().sum()
```

# we found some values are null but fewer than above table

```python
#missing percentage
missing_percentage = data.isnull().sum()*100 / len(data)
missing_percentage
```

# Heatmap To Visualise The Null Presence 

```python
plt.figure(figsize=(6,5))
plt.figure(figsize=(10,10))
sns.heatmap(data.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

# Conclusion Of heatMap

"Longitude" and "Latitude" are also of no use in our analysis so we will remove them also 



# Filling '0' where null value

```python
pd.options.mode.chained_assignment = None  # default='warn'  # to disable chained operation error 
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kills'] = data['kills'].fillna(0)
data['Casualties'] = data['kills'] + data['Wound']
```

# Filling Unknown where there is null value in state and city column

```python
data['State'].fillna('unknown', inplace=True)
data['city'].fillna('unknown', inplace=True)
```

```python
data.head()
```

# Drop all other columns with null values 

```python
data.isna()
```

```python
data.dropna(axis=1, inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

# DATA VISUALIZATION

## CountryWise Attack

```python
plt.rcParams["figure.figsize"]=15, 11
sns.barplot(x = data["Country"].value_counts()[:20].index, y=data["Country"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Region", fontsize=18)
plt.title("Country wise Attacks", size=25, fontweight="bold");
```

# Month Wise attack

```python
plt.rcParams['figure.figsize']= 15, 11
sns.barplot(x = data["Month"].value_counts()[:20].index, y=data["Month"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Month", fontsize=18)
plt.title(" Attacks by Month", size=25, fontweight="bold");
```

# Number of Attacks over different region


```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title("Terrorist Activities by Region In each Year ", fontsize=25)
plt.ylabel("no.of Attacks",fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
pd.crosstab(data.Year, data.Region)
```

# No. of Attacks over Different cities Visualization

```python
data['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color="green")
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("City", fontsize=18)
plt.title(" Top 10 most Affected cities", size=25, fontweight="bold");
```

# Attack Organisations over Attacked cities

```python
plt.rcParams['figure.figsize']= 15, 11
sns.barplot(x = data["Group name"].value_counts()[:20].index, y=data["Group name"].value_counts()[:20].values)
plt.ylabel("Total Attacks", fontsize=18)
plt.xticks(rotation=90)
plt.xlabel("Attack Group", fontsize=18)
plt.title(" Attacks Groups", size=25, fontweight="bold");
```

# Number of kills by each terrorist organisation

```python
group=data[['Group name','Country','kills']]
group = group.groupby(['Group name', 'Country'],axis=0).sum().sort_values('kills',ascending=False).drop('Unknown').reset_index().head(10)
group
```

# Number of deaths over years by Attacks

```python
df=data[['Year','kills']].groupby(['Year']).sum()
fig, ax4= plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation=50)
plt.title("People died due to attack",fontsize=25,fontweight='bold')
plt.ylabel("number of people killed",fontsize=20)
plt.xlabel("Years",fontsize=20)
top_side=ax4.spines["top"]
top_side.set_visible(False)
right_side=ax4.spines["right"]
right_side.set_visible(False)
```

# number of people killed with the type of attack

```python
data[['AttackType','kills']].groupby(['AttackType'],axis=0).sum().plot(kind='bar',figsize=(20,10),color=["darkslateblue"])
plt.xticks(rotation=50)
plt.title("Number of kille",fontsize=25,fontweight='bold')
plt.ylabel("number of people killed",fontsize=20)
plt.xlabel("Attack type",fontsize=20)
plt.show()
```
